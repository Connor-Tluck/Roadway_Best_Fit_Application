## Define all functions used to create features 

```python
def short_bearing(grouped_points):
    i = grouped_points
    num = ((i[1][0]-i[0][0])*(i[2][0]-i[1][0])+(i[1][1]-i[0][1])*(i[2][1]-i[1][1]))
    den = math.sqrt(((i[1][0]-i[0][0])**2)+((i[1][1]-i[0][1])**2))*math.sqrt(((i[2][0]-i[1][0])**2)+((i[2][1]-i[1][1])**2))
    return(math.acos(num/den)*(180/math.pi))

def bearing_finder(grouped_points):
    list_of_thetas = []
    for i in grouped_points:
            num = ((i[1][0]-i[0][0])*(i[2][0]-i[1][0])+(i[1][1]-i[0][1])*(i[2][1]-i[1][1]))
            den = math.sqrt(((i[1][0]-i[0][0])**2)+((i[1][1]-i[0][1])**2))*math.sqrt(((i[2][0]-i[1][0])**2)+((i[2][1]-i[1][1])**2))
            final_angle_calc = math.acos(num/den)*(180/math.pi)
            list_of_thetas.append(final_angle_calc)
    return(list_of_thetas)
def oscillating_circle_radius_finder(x1, y1, x2, y2, x3, y3):
        '''
        Adapted and modifed to get the unknowns for defining a parabola:
        http://stackoverflow.com/questions/717762/how-to-calculate-the-vertex-of-a-parabola-given-three-points
        '''
        denom = (x1-x2) * (x1-x3) * (x2-x3);
        A     = (x3 * (y2-y1) + x2 * (y1-y3) + x1 * (y3-y2)) / denom;
        B     = (x3*x3 * (y1-y2) + x2*x2 * (y3-y1) + x1*x1 * (y2-y3)) / denom;
        C     = (x2 * x3 * (x2-x3) * y1+x3 * x1 * (x3-x1) * y2+x1 * x2 * (x1-x2) * y3) / denom;
            
        p= np.poly1d([A,B,C])
        first_deriv= np.polyder(p,1)
        second_deriv= np.polyder(p,2)
        
        '''Return a list of:
            1. Original Polynomial
            2. First Derivative
            3. Second Derivative'''
        return [p,first_deriv,second_deriv]
    

def cercle_circonscrit(T): 
    (x1, y1), (x2, y2), (x3, y3) = T 
    A = np.array([[x3-x1,y3-y1],[x3-x2,y3-y2]]) 
    Y = np.array([(x3**2 + y3**2 - x1**2 - y1**2),(x3**2+y3**2 - x2**2-y2**2)]) 
    if np.linalg.det(A) == 0: 
        return 0 
    Ainv = np.linalg.inv(A) 
    X = 0.5*np.dot(Ainv,Y) 
    x,y = X[0],X[1] 
    r = sqrt((x-x1)**2+(y-y1)**2) 
#     return (x,y),r 
    return r


def cercle_circonscrit(T): 
    (x1, y1), (x2, y2), (x3, y3) = T 
    A = np.array([[x3-x1,y3-y1],[x3-x2,y3-y2]]) 
    Y = np.array([(x3**2 + y3**2 - x1**2 - y1**2),(x3**2+y3**2 - x2**2-y2**2)]) 
    if np.linalg.det(A) == 0: 
        return 0 
    Ainv = np.linalg.inv(A) 
    X = 0.5*np.dot(Ainv,Y) 
    x,y = X[0],X[1] 
    r = sqrt((x-x1)**2+(y-y1)**2) 
#     return (x,y),r 
    return r



```

```python
'''Processing task to delete any points which are repeated more than once in the dataset, 
this is a cleaning task for multiple shot points which will
cause errors as some calcs might raise errors when dividing by zero. '''

index_of_repeat_points = []
for i in range(len(df)-2):
    if (df.iloc[i]["Northing"] == df.iloc[i+1]["Northing"]) & (df.iloc[i]["Easting"] == df.iloc[i+1]["Easting"]):
        index_of_repeat_points.append(i)
df = df.drop(df.index[index_of_repeat_points])

#reset the index from removing duplicate rows. 
df.reset_index(inplace = True)
df.drop(['index'], axis=1)
```

```python

# for sublist in dfs:
#     slope_to_next_point = [0]
#     for i in range(len(sublist)-1):
#         slope_calcs = (((sublist.iloc[i]["Easting"]),(sublist.iloc[i]["Northing"])),((sublist.iloc[i+1]["Easting"]),(sublist.iloc[i+1]["Northing"])))
#         p1_x=sublist.iloc[i]["Easting"]
#         p1_y=sublist.iloc[i]["Northing"]
#         p2_x=sublist.iloc[i+1]["Easting"]
#         p2_y=sublist.iloc[i+1]["Northing"]
#         slope = (p2_y-p1_y)/(p2_x-p1_x)
#         #remove infinity values
#         if slope == '-inf':
#             slope = 0
#         else:
#             pass
#         slope_to_next_point.append(slope)
#     sublist['slope_to_next_point'] = slope_to_next_point
    

for sublist in dfs:
    bearing_list = [0]
    for i in range(len(sublist)-1):
        try:
            current_segment_bearing = ((sublist.iloc[i]["Easting"]),(sublist.iloc[i]["Northing"])),((sublist.iloc[i+1]["Easting"]),(sublist.iloc[i+1]["Northing"])),((sublist.iloc[i+2]["Easting"]),(sublist.iloc[i+2]["Northing"]))
            bearing_list.append(bearing_finder([current_segment_bearing])[0])
        except:
            bearing_list.append(0)
    sublist['bearing_angles'] = bearing_list

#three point counter
for sublist in dfs:
    id_list = list(sublist['id'].unique())
    counter = 0
    index_list = [0,1,2]
    three_point_sum = [0]
    for index,row in sublist.iterrows():
        sum_list = [sublist.iloc[index_list[0]]['bearing_angles'],sublist.iloc[index_list[1]]['bearing_angles'],sublist.iloc[index_list[2]]['bearing_angles']]
        three_point_sum.append(sum(sum_list))
        index_list = list(map(lambda x : x + 1, index_list)) 
        sum_list = []
        counter += 1
        if counter == len(sublist.index)-2:
            three_point_sum.append(0)
            break
    sublist['three_point_sum'] = three_point_sum
    # print(len(three_point_sum))

#five point counter
for sublist in dfs:
    counter = 0
    index_list = [0,1,2,3,4]
    five_point_sum = [0,0]
    for index,row in sublist.iterrows():
        sum_list = [sublist.iloc[index_list[0]]['bearing_angles'],sublist.iloc[index_list[1]]['bearing_angles'],sublist.iloc[index_list[2]]['bearing_angles'],sublist.iloc[index_list[3]]['bearing_angles'],sublist.iloc[index_list[4]]['bearing_angles']]
        five_point_sum.append(sum(sum_list))
        index_list = list(map(lambda x : x + 1, index_list)) 
        sum_list = []
        counter += 1
        if counter == len(sublist.index)-5:
            five_point_sum.append(0)
            five_point_sum.append(0)
            five_point_sum.append(0)
            break
    sublist['five_point_sum'] = five_point_sum
    # print(len(five_point_sum))

#tangent distance calculator

for sublist in dfs:
    index_list = [0,1]
    tangent_list = []
    counter = 0
    for index,row in sublist.iterrows():
        xb = sublist.iloc[index_list[1]]['Easting_US']
        xa = sublist.iloc[index_list[0]]['Easting_US']
        yb = sublist.iloc[index_list[1]]['Northing_US']
        ya = sublist.iloc[index_list[0]]['Northing_US']
        tangent_length = math.sqrt(((xb-xa)**2) + ((yb-ya)**2))
        tangent_list.append(tangent_length)
        index_list = list(map(lambda x : x + 1, index_list)) 
        lengent_length = []
        counter += 1
        if counter == len(sublist.index)-1:
            tangent_list.append(mean(tangent_list))
            break
    sublist['segment_length'] = tangent_list

#three circonscrit circle radius

for sublist in dfs:
    counter = 0
    index_list = [0,1,2]
    conscrit_list = [0]
    for index,row in sublist.iterrows():
        T = (sublist.iloc[index_list[0]]['Easting_US'],sublist.iloc[index_list[0]]['Northing_US']),(sublist.iloc[index_list[1]]['Easting_US'],sublist.iloc[index_list[1]]['Northing_US']),(sublist.iloc[index_list[2]]['Easting_US'],sublist.iloc[index_list[2]]['Northing_US'])
        circonscrit_circle_r = cercle_circonscrit(T)
        conscrit_list.append(circonscrit_circle_r)
        circonscrit_circle_r = []
        index_list = list(map(lambda x : x + 1, index_list)) 
        counter += 1
        if counter == len(sublist.index)-2:
            conscrit_list.append(0)
            break
    sublist['circonscrit_circle_radius'] = conscrit_list 
    # print(conscrit_list[1][1])

#Oscillating Circle

for sublist in dfs:
    counter = 0
    index_list = [0,1,2]
    oscillating_circle_list = [0]
    for index,row in sublist.iterrows():
        
        #define x of zero.
        x_0 = sublist.iloc[index_list[1]]['Easting_US']
        
        #define the working points for the calcs.
        a1,b1=[sublist.iloc[index_list[0]]['Easting_US'],sublist.iloc[index_list[0]]['Northing_US']]
        a2,b2=[sublist.iloc[index_list[1]]['Easting_US'],sublist.iloc[index_list[1]]['Northing_US']]
        a3,b3=[sublist.iloc[index_list[2]]['Easting_US'],sublist.iloc[index_list[2]]['Northing_US']]
        
        #this variable stores polynomial, derivative 1 and 2.
        polynomial_storage = oscillating_circle_radius_finder(a1, b1, a2, b2, a3, b3)
        
        first_derv_zero = polynomial_storage[1](x_0)
        second_derv_zero = polynomial_storage[2](x_0)
        
        #to avoid an error of div by zero make close to zero.
        if second_derv_zero == 0:
            second_derv_zero = 1.0*10**-13
        else:
            pass

        
        #this solves for the radius of the oscillating circle
        oscillating_circle_r = (1+((first_derv_zero)**2)**(3/2))/(abs(second_derv_zero))

    
        oscillating_circle_list.append(oscillating_circle_r)
        index_list = list(map(lambda x : x + 1, index_list)) 
        counter += 1
#         print(index_list)
#         print(oscillating_circle_r)
        oscillating_circle_r = []


        if counter == len(sublist.index)-2:
            oscillating_circle_list.append(0)
            break
    sublist['oscillating_circle_radius'] = oscillating_circle_list 



```

```python
#Create copy of dataset.
df_model = merged_df.copy()
#Create KNN Object
knn = KNeighborsClassifier()
#Create x and y variable
x = df_model[["bearing_angles",'three_point_sum','five_point_sum','segment_length','circonscrit_circle_radius','oscillating_circle_radius']]
y = df_model['class']
#Split data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)
#Training the model
knn.fit(x_train, y_train)
#Predict testing set
y_pred = knn.predict(x_test)
#Check performance using accuracy
print(accuracy_score(y_test, y_pred))
#Check performance using roc
print(roc_auc_score(y_test, y_pred))

print("Number mislabled points of total %d points : %d"
      %(x_test.shape[0], (y_test != y_pred).sum()))
print("Percentage mislabled points of total %d points : %d PERCENT"
      %(x_test.shape[0], ((y_test != y_pred).sum())/x_test.shape[0]*100))
    
'''export the trained model'''

filename = 'finalized_model.sav'
pickle.dump(knn, open(filename, 'wb'))
```
